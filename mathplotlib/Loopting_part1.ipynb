```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Matplotlib anatomy
### 1.1. Overview
- Một plot hoàn thiện được gọi là một `Figure`
- Mỗi `Figure` chứa một hoặc nhiều `Axes`
- Mỗi `Axes` là một phần diện tích, thể tích mà một điểm (point) có thể được xác định thông qua toạ độ.
    - x,y với 2D plots
    - x,y,z với 3D plots
- Trong phân này ta chỉ quan tâm đến 2D plots
- Mỗi `Axes` chứa các `Axis` (`XAxis` và `YAxis`)
- Mỗi `Axis` chứa các thành phần liên quan đến trục số như ticks, tick locations, labels...
 
### 1.2 Simple example
#### A) Figure

```python
# A blank Figure with no Axes
fig = plt.figure() # red background to see where the figure is
plt.show()
```

#### B) Axes

```python
# A Figure with one blank Axes (most common scenario)
fig, ax = plt.subplots(figsize = (4, 3))
plt.show()
```

#### C) Add Propẻties to `Axes`

```python
# Khởi tạo 1 figure chứa 1 Axes
fig, ax = plt.subplots()

# Set title cho Axes
ax.set_title("This is a title")

# Set limits cho các trục toạ độ 
ax.set_xlim(-5, 5)
ax.set_ylim(0, 5)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.set_title("Nguyễn Quang Huy")

ax.set_xlabel("Weight")
ax.set_ylabel("Height")
ax.set_xlim(-5,5)
ax.set_xlim(7,9)

plt.show()
```

#### D) More customization

```python
# Khởi tạo 1 figure chứa 1 Axes
fig, ax = plt.subplots()

# Set title cho Axes
ax.set_title(
    "This is a title",
    fontsize =20,
    fontweight = "bold",
    fontstyle="italic"
)

# Set limits cho các trục toạ độ
ax.set_xlim(-5,5)
ax.set_ylim(0,4)

ax.set_xlabel("Income", fontsize = 30)
ax.set_ylabel("Education", fontsize = 30)


plt.show()
```

## 2. Two common styles and confusion
- Phần này lấy ví dụ về cùng plot bởi 2 styles: OOP và pylot
- Recommend dùng OOP style
- Phần sau này, các VD đều được trình bày theo OOP style


```python
# Generate data
x = np.linspace(0, 5, 100)
print(len(x))
print(x)
```

## 2.1 OOP style
- Object oriented programming style (recommend)

```python
# Make Figure and Axes
fig, ax = plt.subplots()

# Plot data
ax.plot(x, x, label = "Linear") #line
ax.plot(x, x**2, label = "Quadratic")
ax.plot(x, x**3, label = "Cubic")

#Decorate plot
ax.set_xlabel("X value")
ax.set_ylabel("Y value")
ax.set_title("Simple plot in OOP style")

plt.legend()

plt.show()
```

### Pylot style

```python
plt.plot(x, x, label = "linear")
plt.plot(x, x** 2, label = "Quadratic")
plt.plot(x, x**3, label = "Cubic")

plt.xlabel("X value")
plt.ylabel("Y value")
plt.title("Simple plot in OOP style")

plt.legend()
plt.show()
```

## 3. Matplotlib backend
- Front-end (scripting layer): là code mà user để specify plots
- Backend (backend layer): là cơ chế phía trước mà Matplotlib dùng để gểnates actual plots ứng với từng loại output khác nhau. VD:
    - Inline plots in Jupyter Notebook
    - PDF files
    - PNG files
    - Window popup plots
- Backend quy định cách plot được generate và cách dùng tương tác với plots.
- Có 2 loại backends:
    - User interface backends (or interactive backends): tạo plots trong môi trường interative như jupyter (pygtk, wxpython, tkiner, qt4,...)
    - Hardcopy backends: tạo output file ảnh (png, pdf, svg,...)

## 4. Multi-subplots in one Figure
### VD1

```python
# Tạo 1 figure có 3 Axes
# 1 hàng, 3 cột
fig, ax = plt.subplots(1, 3, figsize = (10, 3))

# Vẽ từng plot vào từng axes
ax[0].plot(x,x)
ax[0].set_title("Nguyen Quang Huy")

ax[1].plot(x, x**2)

ax[2].plot(x, x**3)

plt.show()
```

```python
# Tạo 1 figure có 3 Axes
# 2 hàng, 2 cột
fig, ax = plt.subplots(2, 2, sharey = True)

# Vẽ từng plot vào từng axes
ax[0,0].plot(x,x)
ax[0,0].set_title("Nguyen Quang Huy")

ax[0,1].plot(x, x**2)

ax[1,0].plot(x, x**3)

plt.show()
```

```python
# Tạo 1 figure có 3 Axes
# 2 hàng, 2 cột
fig, ax = plt.subplots(2, 2, sharey = True)

# Vẽ từng plot vào từng axes
ax[0,0].plot(x,x)
ax[0,0].set_title("Nguyen Quang Huy")

ax[0,1].plot(x, x**2)
ax[0,1].set_title("Nguyen Quang Huy")


ax[1,0].plot(x, x**3)
ax[1,0].set_title("Nguyen Quang Huy")

ax[1,1].plot(x, x**3)
ax[1,1].set_title("Nguyen Quang Huy")

# Cách đều các biểu đồ
fig.tight_layout()

plt.show()
```

## 5. Plotting với Pandas
- Thông thường data gặp trong thực tế ở dạng Pandas Series hoặc DataFrame
- Cách plot cũng tương tự như đã học, gồm 4 bước:
    1. Tạo fig, ax
    2. Plot data
    3. Decorate subplots
    4. Save or show

### 5.1 Plot với Series

```python
# Load data
df = sns.load_dataset("tips")
df.head(2)
```

#### VD1: barplot số bữa ăn theo dây

```python
# prepare data
data = df["day"].value_counts()
data
```

```python
# strep 1: fig, ax
fig, axes = plt.subplots(figsize=(5,3))

#step 2:
data.plot.bar(ax = axes)
axes.set_title("Number of meals weekday")
axes.set_ylabel("Number of meals")

plt.show()
```

```python
# strep 1: fig, ax
fig, axes = plt.subplots(figsize=(5,3))

#step 2:
data.plot.barh(ax = axes)
axes.set_title("Number of meals weekday")
axes.set_ylabel("Number of meals")

plt.show()
```

```python
# strep 1: fig, ax
fig, axes = plt.subplots(figsize=(5,3))

#step 2:
data.plot.pie(ax = axes)
axes.set_title("Number of meals weekday")
axes.set_ylabel("Number of meals")

plt.show()
```

### 5.2 Plot với Data Frame

```python
df["sex"].value_counts().plot.bar()
```

### Ví dụ 1: vẽ scatter plot `x = total_bill`, `y=tip`

```python
# Step 1: fig, ax
fig, ax = plt.subplots(figsize = (5,3))

# Step 2: plot data

df.plot.scatter(x = "total_bill", y = "tip", ax = ax)

# step 3: decorate
ax.set_title("Relationship between bill and tip")

plt.show()
```
